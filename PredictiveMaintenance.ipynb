```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(columns='Target')
```

```python
y=df['Target']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
RFC=RandomForestClassifier()
```

```python
LR=LogisticRegression()
```

```python
svm=SVC()
```

```python
DTC=DecisionTreeClassifier()
```

```python
RFC.fit(X_train,y_train)
```

```python
LR.fit(X_train,y_train)
```

```python
svm.fit(X_train,y_train)
```

```python
DTC.fit(X_train,y_train)
```

```python
y_pred1=RFC.predict(X_test)
```

```python
y_pred2=svm.predict(X_test)
```

```python
y_pred3=LR.predict(X_test)
```

```python
y_pred4=DTC.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.metrics import confusion_matrix
```

```python
rfc_accu=accuracy_score(y_test,y_pred1)
```

```python
svm_accu=accuracy_score(y_test,y_pred2)
```

```python
lr_accu=accuracy_score(y_test,y_pred3)
```

```python
dfc_accu=accuracy_score(y_test,y_pred4)
```

```python
print('accuracy of RandomForestClasifier:',rfc_accu)
print('accuracy of svm:',svm_accu)
print('accuracy of LogisticRegression:',lr_accu)
print('accuracy of DecisionTreeClassifier:',dfc_accu)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

precision_rfc = precision_score(y_test, y_pred1)
precision_svm = precision_score(y_test, y_pred2)
precision_lr = precision_score(y_test, y_pred3)
precision_dfc = precision_score(y_test, y_pred4)

print(f" Precision Score For RFC :{precision_rfc}")
print(f" Precision Score For SVM :{precision_svm}")
print(f" Precision Score For LR :{precision_lr}")
print(f" Precision Score For DFC :{precision_dfc}")
```

```python
recall_rfc = recall_score(y_test, y_pred1)
recall_svm = recall_score(y_test, y_pred2)
recall_lr = recall_score(y_test, y_pred3)
recall_dfc = recall_score(y_test, y_pred4)

print(f" Recall Score For RFC :{recall_rfc}")
print(f" Recall Score For SVM :{recall_svm}")
print(f" Recall Score For LR :{recall_lr}")
print(f" Recall Score For DFC :{recall_dfc}")
```

```python
f1_rfc= f1_score(y_test, y_pred1)
f1_svm = f1_score(y_test, y_pred2)
f1_lr = f1_score(y_test, y_pred3)
f1_dfc = f1_score(y_test, y_pred4)

print(f" F1 Score For RFC :{f1_rfc}")
print(f" F1 Score For SVM :{f1_svm}")
print(f" F1 Score For LR :{f1_lr}")
print(f" F1 Score For DFC :{f1_dfc}")
```
